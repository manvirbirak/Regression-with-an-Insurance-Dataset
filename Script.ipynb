```python
import pandas as pd
import numpy as np
import lightgbm as lgb


impute_cols = ['Age', 'Annual Income', 'Health Score', 'Credit Score', 'Number of Dependents']
categorical_cols = ['Number of Dependents']

def lgbm_impute(df, impute_cols, categorical_cols=None):
    data = df.copy()
    data[impute_cols] = data[impute_cols].apply(pd.to_numeric, errors='coerce')
    categorical_cols = categorical_cols or []
    for col in impute_cols:
        mask = data[col].isna()
        if not mask.any():
            continue
        feats = [c for c in impute_cols if c != col]
        X_train = data.loc[~mask, feats].fillna(-9999)
        y_train = data.loc[~mask, col]
        X_pred = data.loc[mask, feats].fillna(-9999)
        if col in categorical_cols:
            model = lgb.LGBMClassifier(n_estimators=100, random_state=42, verbose=-1)
            model.fit(X_train, y_train.astype(int))
            data.loc[mask, col] = model.predict(X_pred).astype(int)
        else:
            model = lgb.LGBMRegressor(n_estimators=100, random_state=42, verbose=-1)
            model.fit(X_train, y_train)
            data.loc[mask, col] = model.predict(X_pred)
    for col in categorical_cols:
        data[col] = data[col].round().astype(int)
    return data

df_imputed = lgbm_impute(train, impute_cols, categorical_cols=['Number of Dependents'])



```

```python
for col in impute_cols:
    train[col] = df_imputed[col]

```
