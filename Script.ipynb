```python
import pandas as pd



train = pd.read_csv("data/train.csv")
test= pd.read_csv("data/test.csv")
```

```python
missing_percentage = train.isnull().mean() * 100
print(missing_percentage)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 5))


train['Occupation'] = train['Occupation'].astype(object)
train['Occupation'].value_counts().plot(kind='bar')
ax.set_title('Occupation Distribution', fontsize=14)
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Occupation', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
train['Occupation'] = train['Occupation'].fillna('NA')
```

```python

fig, ax = plt.subplots(figsize=(6, 5))


train['Previous Claims'] = train['Previous Claims'].astype(object)

train['Previous Claims'].value_counts().plot(kind='bar')
ax.set_title('Previous Claims Distribution', fontsize=14)
ax.set_ylabel('Count', fontsize=12)
ax.set_xlabel('Previous Claims', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(train.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

train.corr(numeric_only=True)['Premium Amount'].sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 5))

train['Credit Score'] = pd.to_numeric(train['Credit Score'], errors='coerce')

sns.kdeplot(train['Credit Score'].dropna(), fill=True, bw_adjust=1)
ax.set_title('Credit Score Distribution', fontsize=14)
ax.set_ylabel(' ', fontsize=12)
ax.set_xlabel('Credit Score', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```
